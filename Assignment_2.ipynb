```python
import numpy as np

array = np.array([1, 2, 3, 4])

sin_result = np.sin(array)
log_result = np.log(array)

log2_result = np.log2(array)

exp_result = np.exp(array)

sin_text = f"np.sin(array) = np.sin({array}) = {sin_result}"
log_text = f"np.log(array) = np.log({array}) = {log_result}"
log2_text = f"np.log2(array) = np.log2({array}) = {log2_result}"
exp_text = f"np.exp(array) = np.exp({array}) = {exp_result}"


print(sin_text)
print(log_text)
print(log2_text)
print(exp_text)
```

Visualisation of data using matplotlib,pyplots packages

```python
import pandas as pd

people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]  # Adjusted length to match people list
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

healthcare_data = pd.DataFrame({
    'Name': people,
    'Age': age,
    'Height (cm)': height,
    'Weight (kg)': weight
})
print(healthcare_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

healthcare_data = pd.DataFrame({
    'Name': people,
    'Age': age,
    'Height (cm)': height,
    'Weight (kg)': weight
})

plt.figure(figsize=(8, 5))
plt.scatter(healthcare_data['Age'], healthcare_data['Weight (kg)'], color='blue')
plt.title('Age vs. Weight')
plt.xlabel('Age (years)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(healthcare_data['Name'], healthcare_data['Height (cm)'], color='orange')
plt.title('Names vs. Height')
plt.xlabel('Name')
plt.ylabel('Height (cm)')
plt.show()
plt.figure(figsize=(8, 5))
plt.hist(healthcare_data['Age'], bins=5, color='purple', edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

healthcare_data = pd.DataFrame({
    'Name': people,
    'Age': age,
    'Height (cm)': height,
    'Weight (kg)': weight
})

plt.figure(figsize=(8, 5))
plt.hist(healthcare_data['Age'], bins=4, color='purple', edgecolor='black')
plt.title('Age Distribution of Individuals in Healthcare Dataset')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()


plt.bar(healthcare_data['Name'], healthcare_data['Height (cm)'], color='orange')
plt.title('Bar Chart of Names vs. Height')
plt.xlabel('Name')
plt.ylabel('Height (cm)')
plt.show()

plt.scatter(healthcare_data['Age'], healthcare_data['Weight (kg)'], color='blue')
plt.title('Scatter Plot of Age vs. Weight')
plt.xlabel('Age (years)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```
